## Project Overview

Financial institutions process millions of transactions every day. Only a small percentage of these transactions can be manually reviewed by fraud analysts due to limited resources. The goal of this project is to build a machine learning model that predicts whether a transaction should be placed into the fraud investigation queue.

The model will assist analysts by prioritizing suspicious transactions while minimizing unnecessary investigations.

# 1. Stakeholder

Primary Stakeholder

• Fraud Investigation Team
• Risk Management Department
• Financial Institution

The machine learning model will assist fraud analysts by identifying transactions that have a high probability of being fraudulent.

The final decision remains with the human analyst.

# 2. Operational Decision

Decision:

Should this transaction be sent to the fraud review queue?

Possible Outcomes

YES
→ Transaction is reviewed by a fraud analyst.

NO
→ Transaction is processed normally.

# 3. Unit of Analysis

One row in the dataset represents:

One individual financial transaction.

Each transaction contains:

- Transaction ID
- Transaction Amount
- Merchant Category
- Transaction Hour
- Device Trust Score
- Velocity of transactions in the last 24 hours
- Cardholder Age
- Fraud Label

# 4. Machine Learning Task

Task Type:

Binary Classification

Target Variable:

is_fraud

Target Classes

0 → Legitimate Transaction

1 → Fraudulent Transaction

The objective is to classify each transaction into one of these two categories.

# 5. Prediction Target

The model predicts whether a transaction should enter the fraud review queue.

Prediction Horizon

Immediately after the transaction is received.

Information Available at Prediction Time

The following information is available before making the prediction:

- Transaction Amount
- Transaction Hour
- Merchant Category
- Foreign Transaction Indicator
- Location Mismatch
- Device Trust Score
- Velocity in Last 24 Hours
- Cardholder Age

The fraud label (is_fraud) is NOT available during prediction because it is the value being predicted.

# 6. Cost of Prediction Errors

False Positive (FP)

A legitimate transaction is flagged as fraud.

Cost:

- Unnecessary analyst investigation
- Increased operational cost
- Customer inconvenience
- Delays in transaction processing

False Negative (FN)

A fraudulent transaction is classified as legitimate.

Cost:

- Financial loss
- Chargebacks
- Reputational damage
- Customer dissatisfaction

Non-use

If no machine learning system is used:

- Analysts manually inspect transactions
- Fraudulent transactions may be missed
- Investigation resources are inefficiently allocated

# 7. Constraints

Business Constraints

- Limited fraud investigation team
- Limited review capacity
- High class imbalance
- Need for fast predictions

Technical Constraints

- Real-time prediction
- Limited computational resources
- Model should be explainable

Ethical Constraints

- Avoid discrimination
- Protect customer privacy
- Human oversight must remain in place

# 8. Harm Analysis

Potential Harms

1. Incorrectly blocking legitimate transactions

2. Missing fraudulent transactions

3. Bias against specific customer groups

4. Reduced customer trust

Mitigation

- Human review of flagged transactions
- Regular model monitoring
- Periodic retraining
- Performance evaluation on new data

# 9. Success Criteria

The model will be considered successful if it:

- Achieves high Recall for fraud detection.
- Maintains acceptable Precision.
- Improves analyst efficiency.
- Reduces missed fraud cases.

Target Metrics

Recall ≥ 90%

Precision ≥ 80%

F1 Score ≥ 85%

ROC-AUC ≥ 0.90

# 10. Baseline

Non-ML Baseline

Rule:

Review every foreign transaction.

or

Review every transaction above a specified amount.

This rule-based system serves as the baseline for comparison with machine learning models.

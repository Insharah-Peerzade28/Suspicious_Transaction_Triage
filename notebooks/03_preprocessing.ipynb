# 03 - Data Preprocessing

## Objective

The objective of preprocessing is to prepare the dataset for machine learning.

The preprocessing steps include:

- Loading the cleaned dataset
- Checking missing values
- Removing duplicates
- Converting categorical variables
- Separating features and target
- Splitting the dataset
- Scaling numerical features
- Preventing data leakage using a preprocessing pipeline

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

from sklearn.compose import ColumnTransformer

from sklearn.pipeline import Pipeline

print("Libraries Loaded Successfully")
```

```python
df = pd.read_csv("../data/raw/credit_card_fraud_10k.csv")

df.head()
```

```python
print("Dataset Shape:", df.shape)

df.info()
```

```python
print(df.isnull().sum())
```

```python
# Numerical Columns

numerical_columns = [
    "amount",
    "transaction_hour",
    "device_trust_score",
    "velocity_last_24h",
    "cardholder_age"
]

for col in numerical_columns:
    df[col] = df[col].fillna(df[col].median())

# Categorical Columns

categorical_columns = [
    "merchant_category",
    "foreign_transaction",
    "location_mismatch"
]

for col in categorical_columns:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Missing values handled successfully.")
```

```python
print("Duplicates Before:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicates After:", df.duplicated().sum())
```

```python
df = df.drop(columns=["transaction_id"])

df.head()
```

```python
X = df.drop("is_fraud", axis=1)

y = df["is_fraud"]

print(X.shape)
print(y.shape)
```

```python
numeric_features = [
    "amount",
    "transaction_hour",
    "device_trust_score",
    "velocity_last_24h",
    "cardholder_age"
]

categorical_features = [
    "merchant_category",
    "foreign_transaction",
    "location_mismatch"
]

print("Numerical Features")

print(numeric_features)

print()

print("Categorical Features")

print(categorical_features)
```

```python
preprocessor = ColumnTransformer(

    transformers=[

        ("num", StandardScaler(), numeric_features),

        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features)

    ]

)

print(preprocessor)
```

```python
X_train, X_test, y_train, y_test = train_test_split(

    X,

    y,

    test_size=0.20,

    random_state=42,

    stratify=y

)

print("Training Set:", X_train.shape)

print("Testing Set :", X_test.shape)
```

```python
X_train_processed = preprocessor.fit_transform(X_train)

X_test_processed = preprocessor.transform(X_test)

print("Training Data Processed")

print("Testing Data Processed")
```

```python
processed_df = pd.concat(
    [
        X.reset_index(drop=True),
        y.reset_index(drop=True)
    ],
    axis=1
)

processed_df.to_csv(
    "../data/processed/processed_transactions.csv",
    index=False
)

print("Processed dataset saved successfully.")
```

```python
import joblib

joblib.dump(
    preprocessor,
    "../models/pipeline.pkl"
)

print("Pipeline Saved Successfully")
```

# Leakage Check

The following precautions were taken to prevent data leakage:

- Transaction ID was removed before training.
- Train-test split was performed before fitting the preprocessing pipeline.
- Scaling was fitted only on the training set.
- One-Hot Encoding was fitted only on the training set.
- The target variable (`is_fraud`) was excluded from preprocessing.
